```python
#bunu en bastan kaldiricaz
DataFrameAraba=DataFrameAraba.drop("transmission",axis=1)
DataFrameAraba
```

```python
#Datada ögrnecek kolonları ve sonuc kolonlarını ayırıyoruz
y=DataFrameAraba["price"].values
x=DataFrameAraba.drop("price",axis=1).values
```

```python
#Verileri train ve test data olarak 2 ye ayırıyoruz
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Makinin daha rahat ögrenmesi icin verileri 0 ile 1 arasında skaler ediyoruz
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Modelin insaasina basliyoruz
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")

```

```python
model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
#Kayip verilerine bakıyoruz ne kadar dusuk olursa okadar iyi
KayipVeri=pd.DataFrame(model.history.history)
KayipVeri.head()
```

```python
#Kayip verileri görselestiiryoruz
KayipVeri.plot()
```

```python
#Test datasını sokup test etiriyoruz
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
TahminiFrame=model.predict(x_test)
```

```python
#Hata oranına bakıyoruz
```

```python
mean_absolute_error(y_test,TahminiFrame)
```
